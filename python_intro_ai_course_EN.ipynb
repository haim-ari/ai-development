````python
# === While Loops ===

# retry loop — very common pattern in AI APIs
import time

max_retries = 3
attempt = 0

while attempt < max_retries:
    attempt += 1
    print(f"Attempt {attempt}/{max_retries}...")
    
    # Simulate success on third attempt
    success = (attempt == 3)
    
    if success:
        print("Request succeeded!")
        break
    else:
        wait_time = 2 ** attempt  # Exponential backoff
        print(f"  Failed. Retrying in {wait_time}s...")
        # time.sleep(wait_time)  # In practice we'd wait, skipping here
else:
    # else on while runs only if loop finished without break
    print("All retries failed!")
````

## 📌 Part 6: Comprehensions — Smart Shortcuts

Comprehensions are a Pythonic and concise way to create lists, dictionaries, and sets.  
They are **faster** than regular loops and **more readable** (when kept short).


````python
# === List Comprehension ===

# Instead of:
numbers = []
for i in range(10):
    numbers.append(i ** 2)

# Write:
numbers = [i ** 2 for i in range(10)]
print(f"Squares: {numbers}")

# With condition (filtering)
even_squares = [i ** 2 for i in range(10) if i % 2 == 0]
print(f"Even squares: {even_squares}")

# AI example: process a list of documents
documents = ["  Hello World  ", "", "Python is great", "  ", "AI rocks!"]
cleaned = [doc.strip() for doc in documents if doc.strip()]
print(f"Cleaned docs: {cleaned}")

# === Dict Comprehension ===
models = ["gpt-4o", "claude-sonnet", "gemini-pro"]
costs = [0.01, 0.003, 0.005]
price_map = {model: cost for model, cost in zip(models, costs)}
print(f"\nPrice map: {price_map}")

# Filter dictionary
cheap_models = {k: v for k, v in price_map.items() if v < 0.005}
print(f"Cheap models: {cheap_models}")

# === Set Comprehension ===
words = ["hello", "HELLO", "Hello", "world", "World"]
unique_lower = {w.lower() for w in words}
print(f"\nUnique words: {unique_lower}")
````

## 📌 Part 7: Functions

Functions are the most basic building block of any program.  
In AI development, functions allow us to create a **modular pipeline** — each step is a function.


````python
# === Basic Functions ===

def greet(name: str) -> str:
    """Greet the user."""
    return f"Hello {name}! Welcome to the AI Course"

print(greet("Student"))

# Parameters with defaults
def call_llm(
    prompt: str,
    model: str = "claude-sonnet-4-20250514",
    temperature: float = 0.7,
    max_tokens: int = 1024
) -> dict:
    """Simulate an LLM API call."""
    return {
        "model": model,
        "prompt": prompt[:50] + "..." if len(prompt) > 50 else prompt,
        "temperature": temperature,
        "max_tokens": max_tokens,
        "response": f"[Simulated response for: {prompt[:30]}...]"
    }

# Different calls
result1 = call_llm("What is Python?")
result2 = call_llm("Explain AI", model="gpt-4o", temperature=0.3)

print(f"\nResult 1: {result1}")
print(f"Result 2: {result2}")
````

````python
# === *args and **kwargs ===

# *args — unknown number of positional arguments
def merge_responses(*responses: str) -> str:
    """Merge multiple responses into one."""
    return " | ".join(responses)

print(merge_responses("Answer 1", "Answer 2", "Answer 3"))

# **kwargs — unknown number of keyword arguments
def create_message(**kwargs) -> dict:
    """Create a message structure from any parameters passed."""
    return {k: v for k, v in kwargs.items()}

msg = create_message(role="user", content="Hello!", name="Student")
print(f"\nMessage: {msg}")

# Combined — common pattern in API wrappers
def api_call(endpoint: str, method: str = "POST", *args, **kwargs):
    print(f"Calling {method} {endpoint}")
    print(f"  Extra args: {args}")
    print(f"  Extra kwargs: {kwargs}")

api_call("/v1/messages", "POST", "extra1", model="claude", stream=True)
````

````python
# === Lambda — Anonymous Functions ===

# Useful for sorting and filtering
models = [
    {"name": "GPT-4o", "cost": 0.01, "score": 92},
    {"name": "Claude", "cost": 0.003, "score": 95},
    {"name": "Gemini", "cost": 0.005, "score": 88},
]

# Sort by score (highest first)
by_score = sorted(models, key=lambda m: m["score"], reverse=True)
print("Sorted by score:")
for m in by_score:
    print(f"  {m['name']}: {m['score']}%")

# Sort by cost (lowest first)
by_cost = sorted(models, key=lambda m: m["cost"])
print("\nSorted by cost:")
for m in by_cost:
    print(f"  {m['name']}: ${m['cost']}")

# map + lambda
names = list(map(lambda m: m["name"].upper(), models))
print(f"\nModel names: {names}")

# filter + lambda
cheap = list(filter(lambda m: m["cost"] < 0.005, models))
print(f"Cheap models: {[m['name'] for m in cheap]}")
````

## 📌 Part 8: Error Handling

In AI development, errors are **part of the routine**:
- API returns an error (rate limit, timeout)
- Model returns invalid JSON
- File not found
- Response missing fields

You **must** write code that handles errors gracefully!


````python
# === Try / Except / Finally ===

# Basic
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")

# Catching specific errors
def parse_api_response(response_text: str) -> dict:
    """Parse an API response — demonstrates error handling."""
    import json
    
    try:
        data = json.loads(response_text)
        content = data["choices"][0]["message"]["content"]
        return {"status": "success", "content": content}
    
    except json.JSONDecodeError as e:
        return {"status": "error", "error": f"Invalid JSON: {e}"}
    
    except KeyError as e:
        return {"status": "error", "error": f"Missing field: {e}"}
    
    except IndexError:
        return {"status": "error", "error": "Empty choices array"}
    
    except Exception as e:
        return {"status": "error", "error": f"Unexpected: {type(e).__name__}: {e}"}
    
    finally:
        print("  [parse_api_response completed]")

# Test with different inputs
test_cases = [
    '{"choices": [{"message": {"content": "Hello!"}}]}',  # Valid
    'not json at all',                                       # Invalid JSON
    '{"choices": []}',                                       # Empty list
    '{"data": "wrong structure"}',                           # Wrong structure
]

for i, test in enumerate(test_cases):
    print(f"\nTest {i+1}:")
    result = parse_api_response(test)
    print(f"  Result: {result}")
````

````python
# === Raising Exceptions ===

def validate_temperature(temp: float) -> float:
    """Validate that temperature is in valid range."""
    if not isinstance(temp, (int, float)):
        raise TypeError(f"Temperature must be a number, got {type(temp).__name__}")
    if not 0.0 <= temp <= 2.0:
        raise ValueError(f"Temperature must be 0.0-2.0, got {temp}")
    return float(temp)

# Tests
try:
    print(validate_temperature(0.7))   # Valid
    print(validate_temperature(3.0))   # Out of range
except ValueError as e:
    print(f"Validation error: {e}")

try:
    print(validate_temperature("hot"))  # Wrong type
except TypeError as e:
    print(f"Type error: {e}")
````

## 📌 Part 9: Object-Oriented Programming (OOP)

OOP lets us organize code into **classes** — templates for creating objects.

In the AI world, almost every framework uses OOP:
- LangChain: `ChatOpenAI()`, `PromptTemplate()`, `VectorStore()`
- FastAPI: `app = FastAPI()`
- Pydantic: `class Config(BaseModel)`


````python
# === Basic Class ===

class LLMClient:
    """Simple client for communicating with a language model."""
    
    # Class variable — shared across all instances
    DEFAULT_MODEL = "claude-sonnet-4-20250514"
    
    def __init__(self, api_key: str, model: str = None):
        """Constructor — called when creating a new instance."""
        self.api_key = api_key           # Instance variable
        self.model = model or self.DEFAULT_MODEL
        self.history: list[dict] = []    # Conversation history
        self._request_count = 0          # private (convention)
    
    def chat(self, message: str) -> str:
        """Send a message and get a response (simulation)."""
        self._request_count += 1
        self.history.append({"role": "user", "content": message})
        
        # Simulate response
        response = f"[{self.model}] Response to: {message[:30]}..."
        self.history.append({"role": "assistant", "content": response})
        
        return response
    
    def get_stats(self) -> dict:
        """Return usage statistics."""
        return {
            "model": self.model,
            "requests": self._request_count,
            "history_length": len(self.history)
        }
    
    def __repr__(self) -> str:
        return f"LLMClient(model='{self.model}', requests={self._request_count})"

# Usage
client = LLMClient(api_key="sk-test123")
print(client.chat("What is Python?"))
print(client.chat("How do I use classes?"))
print(f"\nStats: {client.get_stats()}")
print(f"Repr: {client}")
````

````python
# === Inheritance ===

class StreamingLLMClient(LLMClient):
    """Client with streaming support."""
    
    def __init__(self, api_key: str, model: str = None, chunk_size: int = 10):
        super().__init__(api_key, model)  # Call parent constructor
        self.chunk_size = chunk_size
    
    def stream_chat(self, message: str):
        """Send a message and return response in chunks (simulation)."""
        full_response = self.chat(message)  # Use parent's function
        
        # Simulate streaming
        words = full_response.split()
        for i in range(0, len(words), 2):
            chunk = " ".join(words[i:i+2])
            yield chunk  # Generator — returns one chunk at a time!

# Usage
streaming_client = StreamingLLMClient(api_key="sk-test", chunk_size=5)
print("Streaming response:")
for chunk in streaming_client.stream_chat("Tell me about AI"):
    print(f"  chunk: '{chunk}'")

print(f"\nStats: {streaming_client.get_stats()}")
print(f"Is LLMClient? {isinstance(streaming_client, LLMClient)}")
````

````python
# === Context Manager — the with statement ===

# 'with' ensures automatic resource cleanup

# Custom Context Manager example
class Timer:
    """Measures execution time of a code block."""
    
    def __enter__(self):
        import time
        self.start = time.time()
        return self
    
    def __exit__(self, exc_type, exc_val, exc_tb):
        import time
        self.elapsed = time.time() - self.start
        print(f"Elapsed: {self.elapsed:.4f} seconds")
        return False  # Don't swallow errors

# Usage
with Timer():
    # Simulate processing
    total = sum(i ** 2 for i in range(100_000))
    print(f"Sum: {total:,}")
````

## 📌 Part 13: Generators — Lazy Data Production

Generators are functions that produce values **one at a time** — saving memory!  
In AI, streaming responses are based on generators.


````python
# === Generator Function ===

def token_stream(text: str, chunk_size: int = 3):
    """Simulate streaming — return words in groups."""
    words = text.split()
    for i in range(0, len(words), chunk_size):
        chunk = " ".join(words[i:i + chunk_size])
        yield chunk  # yield instead of return!

# Usage — each chunk is produced only when needed
response = "Python is an amazing programming language that is widely used for AI development"

print("Streaming response:")
for chunk in token_stream(response, chunk_size=3):
    print(f"  >> {chunk}")

# === Generator Expression (one-liner) ===
# Just like list comprehension, but with () instead of []
numbers = range(1_000_000)
sum_of_squares = sum(x**2 for x in numbers)  # Doesn't create a list in memory!
print(f"\nSum of squares (1M numbers): {sum_of_squares:,}")

# === next() — manually get the next value ===
gen = token_stream("Hello World from Python AI")
print(f"\nFirst chunk: {next(gen)}")
print(f"Second chunk: {next(gen)}")
````

## 📌 Part 14: Decorators — Function Wrappers

Decorators add functionality **around** an existing function — without changing it.  
Very common in Python, especially in FastAPI, Flask, pytest.


````python
# === Basic Decorator — Timer ===
import time
import functools

def timer(func):
    """Measure function execution time."""
    @functools.wraps(func)  # Preserve original function's name and docstring
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        elapsed = time.time() - start
        print(f"{func.__name__} took {elapsed:.4f}s")
        return result
    return wrapper

@timer
def process_documents(docs: list[str]) -> list[str]:
    """Process a list of documents."""
    return [doc.upper().strip() for doc in docs]

# Calling process_documents will automatically print execution time
result = process_documents(["hello world", "  python  ", "ai course"])
print(f"Result: {result}")

# === Decorator with Parameters — Retry ===
def retry(max_attempts: int = 3, delay: float = 1.0):
    """Decorator that retries on failure."""
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            for attempt in range(1, max_attempts + 1):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    print(f"  Attempt {attempt}/{max_attempts} failed: {e}")
                    if attempt == max_attempts:
                        raise
                    # time.sleep(delay)  # In practice we'd wait
            return None
        return wrapper
    return decorator

@retry(max_attempts=3, delay=0.5)
def flaky_api_call(url: str) -> str:
    """Simulate an API call that sometimes fails."""
    import random
    if random.random() < 0.6:  # 60% chance of failure
        raise ConnectionError("API timeout")
    return f"Success from {url}"

print("\nCalling flaky API:")
try:
    result = flaky_api_call("https://api.example.com")
    print(f"Result: {result}")
except ConnectionError:
    print("All attempts failed!")
````

## 📌 Part 18: Async/Await — Asynchronous Programming

When working with APIs, a lot of time is wasted **waiting for responses**.  
`async/await` allows sending requests **in parallel** — many times faster!

> Note: In Jupyter, `asyncio` works slightly differently. We'll use `await` directly.


````python
# === Async Basics ===
import asyncio

async def fetch_model_response(model_name: str, delay: float) -> dict:
    """Simulate an async API call."""
    print(f"  Calling {model_name}...")
    await asyncio.sleep(delay)  # Simulate waiting
    print(f"  {model_name} responded!")
    return {"model": model_name, "response": f"Hello from {model_name}"}

async def main():
    # === Sequential - one after another (slow!) ===
    import time
    
    start = time.time()
    r1 = await fetch_model_response("GPT-4o", 1.0)
    r2 = await fetch_model_response("Claude", 0.8)
    r3 = await fetch_model_response("Gemini", 1.2)
    seq_time = time.time() - start
    print(f"Sequential: {seq_time:.1f}s\n")
    
    # === Parallel - at the same time (fast!) ===
    start = time.time()
    results = await asyncio.gather(
        fetch_model_response("GPT-4o", 1.0),
        fetch_model_response("Claude", 0.8),
        fetch_model_response("Gemini", 1.2),
    )
    par_time = time.time() - start
    print(f"Parallel: {par_time:.1f}s")
    print(f"Speedup: {seq_time/par_time:.1f}x faster!")
    
    return results

# In Jupyter, you can simply: await main()
# In a regular script:
results = asyncio.run(main())
````

## 📌 Part 19: Useful Patterns for AI Development

### Design patterns you'll see again and again in AI code:


````python
# === Pattern 1: Configuration with Defaults ===

from dataclasses import dataclass

@dataclass
class PipelineConfig:
    """Pipeline configuration with smart defaults."""
    model: str = "claude-sonnet-4-20250514"
    temperature: float = 0.7
    max_tokens: int = 2048
    chunk_size: int = 500
    chunk_overlap: int = 50
    top_k: int = 5
    system_prompt: str = "You are a helpful assistant"
    
    def __post_init__(self):
        """Validation after creation."""
        assert 0 <= self.temperature <= 2, "Temperature must be 0-2"
        assert self.chunk_overlap < self.chunk_size, "Overlap must be < chunk size"

config = PipelineConfig(temperature=0.3, top_k=10)
print(f"Config: {config}")
````

````python
# === Pattern 2: Batch Processing with Progress ===

def process_in_batches(items: list, batch_size: int = 5) -> list:
    """Process items in batches - essential for embeddings and API calls."""
    results = []
    total = len(items)
    
    for i in range(0, total, batch_size):
        batch = items[i:i + batch_size]
        batch_num = i // batch_size + 1
        total_batches = (total + batch_size - 1) // batch_size
        
        print(f"  Processing batch {batch_num}/{total_batches} "
              f"({len(batch)} items)...")
        
        # "Process" the batch
        processed = [item.upper() for item in batch]
        results.extend(processed)
    
    return results

documents = [f"doc_{i}" for i in range(23)]
results = process_in_batches(documents, batch_size=5)
print(f"\nProcessed {len(results)} documents")
````

````python
# === Pattern 3: Simple Text Chunking (RAG Foundation) ===

def chunk_text(
    text: str,
    chunk_size: int = 200,
    overlap: int = 50
) -> list[dict]:
    """Split text into chunks with overlap - the foundation of RAG!"""
    chunks = []
    start = 0
    chunk_id = 0
    
    while start < len(text):
        end = start + chunk_size
        chunk_text = text[start:end]
        
        chunks.append({
            "id": f"chunk_{chunk_id}",
            "text": chunk_text,
            "start": start,
            "end": min(end, len(text)),
            "length": len(chunk_text)
        })
        
        chunk_id += 1
        start += chunk_size - overlap  # Overlap!
    
    return chunks

# Example
sample_text = (
    "Python is a versatile programming language. "
    "It is widely used in artificial intelligence and machine learning. "
    "Python's simple syntax makes it easy to learn. "
    "Libraries like NumPy, Pandas, and PyTorch power the AI ecosystem. "
    "Python is also great for web development with frameworks like FastAPI and Django."
)

chunks = chunk_text(sample_text, chunk_size=100, overlap=20)
print(f"Text length: {len(sample_text)} chars")
print(f"Number of chunks: {len(chunks)}\n")
for chunk in chunks:
    print(f"[{chunk['id']}] ({chunk['start']}-{chunk['end']}, {chunk['length']} chars)")
    print(f"  '{chunk['text'][:60]}...'\n")
````

````python
# === Pattern 4: Simple Prompt Template ===

class PromptTemplate:
    """Simple prompt template - like LangChain, but basic."""
    
    def __init__(self, template: str):
        self.template = template
    
    def format(self, **kwargs) -> str:
        """Fill in template variables."""
        try:
            return self.template.format(**kwargs)
        except KeyError as e:
            raise ValueError(f"Missing variable in template: {e}")
    
    def get_variables(self) -> list[str]:
        """Return the variable names in the template."""
        import re
        return re.findall(r'\{(\w+)\}', self.template)

# Usage
qa_template = PromptTemplate(
    template="""Answer the question based on the context below.

Context: {context}

Question: {question}

Answer in {language}. Be concise and accurate."""
)

print(f"Variables: {qa_template.get_variables()}")
print()

prompt = qa_template.format(
    context="Python was created by Guido van Rossum in 1991.",
    question="When was Python created?",
    language="English"
)
print(prompt)
````
